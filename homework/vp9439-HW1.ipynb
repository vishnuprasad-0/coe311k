```python
import numpy as np
import matplotlib.pyplot as plt

samples = np.arange(10,10000,20)
error = np.zeros(len(samples))

for n in samples:
    i = int((n-10)/20)
    estPi = monteCarlo(n)
    error[i] = np.abs(np.pi - estPi)
    
plot()
```

```python
def monteCarlo(n):
    x = np.random.rand(n)
    y = np.random.rand(n)
    z = x**2 + y**2
    z = z[z < 1]
    estPi = 4 * len(z) / n
    return estPi
```

```python
def plot():
    plt.plot(samples, error, 'b.')
    plt.xlabel("Number of Samples")
    plt.ylabel("Error")
    plt.show()
```
