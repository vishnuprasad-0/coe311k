```python
import numpy as np
import matplotlib.pyplot as plt
import math

inputs = np.arange(0,10.1,0.1)
error = np.zeros(4)

for x in inputs:
    error[0] = np.abs(np.sin(x) - sin(x,1))
    error[1] = np.abs(np.sin(x) - sin(x,2))
    error[2] = np.abs(np.sin(x) - sin(x,3))
    error[3] = np.abs(np.sin(x) - sin(x,4))
    plot(error,x)
```

## For smaller values of x, truncation error decreases uniformly as the number of terms increase. As x gets larger, truncation error gets larger and sometimes increases when more terms are added. At the largest values of x, truncation error is extremely large and increases uniformly as the number of terms increase.

```python
def sin(x,n):
    output = 0
    for i in range(n):
        output += (((-1)**i)*(x**(2*i+1)))/math.factorial(2*i+1)
    return output
```

```python
def plot(error,x):
    plt.plot(range(1,5), error, 'b.')
    plt.xlabel('Number of Terms')
    plt.ylabel('Truncation Error')
    plt.title(f'Truncation Error vs. Number of Terms (x = {int(x*10)/10})')
    plt.show()
```
