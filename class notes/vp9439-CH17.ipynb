```python
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
k = 1.0        # Constant in the equation (e.g., spring constant)
h = 0.01       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx

def defineParameters(h):
    # Create arrays to store x, y, and v values
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0
    
    return x_values, y_values, v_values

stepSize = np.array([.01, .02, .03, .05, .1, .2, .3, .5])
plt.figure(figsize=(10, 5))
for h in stepSize:
    x_values, y_values, v_values = defineParameters(h)
    
    # Euler's Method loop
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    
        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
    plt.plot(x_values, y_values, label= f'Position y(x): h = {h}')
    plt.plot(x_values, v_values, label= f'Velocity v(x): h = {h}', linestyle='--')

# Plot the results
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```
