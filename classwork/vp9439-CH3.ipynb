```python
import numpy as np
import matplotlib.pyplot as plt

absoluteError = np.zeros(100)
relativeError = np.zeros(100)

for i in range(100):
    estPi = monteCarlo(1000)
    absoluteError[i] = np.abs(estPi - np.pi)
    relativeError[i] = np.abs((estPi - np.pi) / np.pi)

plot()
```

```python
def monteCarlo(n):
    x = np.random.rand(n)
    y = np.random.rand(n)
    z = x**2 + y**2
    z = z[z < 1]
    estPi = 4 * len(z) / n
    return estPi
```

```python
def plot():
    plt.plot(range(100), absoluteError, 'b.', label = "Absolute Error")
    plt.plot(range(100), relativeError, 'r.', label = "Relative Error")
    plt.xlabel("Trial")
    plt.ylabel("Error")
    plt.legend()
    plt.show()
```
