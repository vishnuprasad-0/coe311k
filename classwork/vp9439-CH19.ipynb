```python
import numpy as np
import matplotlib.pyplot as plt

t_start, t_end = 0, 30

def defineParameters(h):
    num_steps = int((t_end - t_start) / h)
    t_values = np.linspace(t_start, t_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 0 # y(0) = 0
    return num_steps, t_values, y_values

# Function representing dy/dx = -2y
def f(t, y):
    return -1000*y + 3000 - 2000*np.exp(-t)

def exact_solution(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)

h_values = np.array([.1, .01, .001, .00001, .000001])
for h in h_values:
    num_steps, t_values, y_values = defineParameters(h)

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])
    plt.plot(t_values, y_values, label=f'Euler\'s Method (h={h})')

t_exact = np.linspace(t_start, t_end, 1000)
y_exact = exact_solution(t_exact)
plt.plot(t_exact, y_exact, 'r-', label='Exact Solution')

plt.xlabel('t')
plt.ylabel('y')
plt.title('Stiff Equation using Euler Forward')
plt.legend()
plt.grid(True)
plt.show()
```
