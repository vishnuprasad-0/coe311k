```python
import numpy as np
import matplotlib.pyplot as plt

x_start, x_end = 0, 10

def defineParameters(h):
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1 # y(0) = 1
    return num_steps, x_values, y_values

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

# Exact function
def exactf(x, y):
    return np.exp(-2 * x)

def calcError(x, y):
    y_exact = exactf(x, y)
    localError = y_exact - y
    globalError = np.sum(np.abs(localError))
    return localError, globalError

def plot(h):
    plt.plot(x_values, y_values, label=f'h = {h:.2f}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f"Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()

stepSize = np.array([.1, .5, 1, 1.5])
for h in stepSize:
    num_steps, x_values, y_values = defineParameters(h)
    
    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    localError, globalError = calcError(x_values, y_values)
    plot(h)
    print(f'Local Error: {localError}')
    print(f'Global Error: {globalError}')
```
