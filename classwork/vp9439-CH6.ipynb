```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]], dtype = float)
b = np.array([14, 30, 26, 25, 37], dtype = float)
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100

start = time.time()
solution_jacobi, iterations_jacobi = jacobi(A, b, x0, tol, max_iterations)
end = time.time()
print(f"Jacobi Solution: {solution_jacobi}")
if iterations_jacobi < max_iterations:
    print(f"Jacobi Iterations: {iterations_jacobi}")
    print(f"Elapsed Time: {end - start} sec")
else:
    print("Jacobi Method diverges.\n")

start = time.time()
solution_gauss, iterations_gauss = gauss_seidel(A, b, x0, tol, max_iterations)
end = time.time()
print(f"Gauss-Seidel Solution: {solution_gauss}")
if iterations_gauss < max_iterations:
    print(f"Gauss-Seidel Iterations: {iterations_gauss}")
    print(f"Elapsed Time: {end - start} sec")
else:
    print("Gauss-Seidel method diverges.")
```
