```python
import numpy as np
import matplotlib.pyplot as plt
import time

A1 = np.random.rand(3,3)
A2 = np.random.rand(3,3)

B1 = np.random.rand(6,6)
B2 = np.random.rand(6,6)

C1 = np.random.rand(12,12)
C2 = np.random.rand(12,12)

D1 = np.random.rand(24,24)
D2 = np.random.rand(24,24)

E1 = np.random.rand(48,48)
E2 = np.random.rand(48,48)

F1 = np.random.rand(96,96)
F2 = np.random.rand(96,96)

G1 = np.random.rand(192,192)
G2 = np.random.rand(192,192)

H1 = np.random.rand(384,384)
H2 = np.random.rand(384,384)
```

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                
    return C
```

```python
def calcRuntime(X, Y):
    start = time.time()
    Z = naive_matrix_multiplication(X, Y)
    end = time.time()
    runtime1 = end - start
    
    start = time.time()
    Z = np.dot(X, Y)
    end = time.time()
    runtime2 = end - start
    
    return runtime1, runtime2
```

```python
runtimeNaive = np.zeros(8)
runtimeOptimized = np.zeros(8)

runtimeNaive[0], runtimeOptimized[0] = calcRuntime(A1, A2)
runtimeNaive[1], runtimeOptimized[1] = calcRuntime(B1, B2)
runtimeNaive[2], runtimeOptimized[2] = calcRuntime(C1, C2)
runtimeNaive[3], runtimeOptimized[3] = calcRuntime(D1, D2)
runtimeNaive[4], runtimeOptimized[4] = calcRuntime(E1, E2)
runtimeNaive[5], runtimeOptimized[5] = calcRuntime(F1, F2)
runtimeNaive[6], runtimeOptimized[6] = calcRuntime(G1, G2)
runtimeNaive[7], runtimeOptimized[7] = calcRuntime(H1, H2)

matNumber = np.array([3, 6, 12, 24, 48, 96, 192, 384])
```

```python
plt.plot(matNumber, runtimeNaive, 'r.', label = 'Naive Runtimes')
plt.plot(matNumber, runtimeOptimized, 'b.', label = 'Optimized Runtimes')
plt.xlabel('Matrix Number')
plt.ylabel('Runtime')
plt.legend()
plt.show()
```
