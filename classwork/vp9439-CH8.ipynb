```python
import numpy as np

def isDiagonallyDominant(A):
    n = len(A[0])
    for i in range(n):
        rowSum = 0
        for j in range(n):
            if i != j:
                rowSum += np.abs(A[i,j])
        if np.abs(A[i,i]) < rowSum:
            return False
    return True
```

```python
A = np.array([[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
              [0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
              [0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
              [0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
              [0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)
print(A)
print(f"Is the matrix diagonally dominant? {isDiagonallyDominant(A)}")
```
