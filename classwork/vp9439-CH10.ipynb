```python
import numpy as np
import matplotlib.pyplot as plt

def evaluateFit(y, y_pred):
    # Mean Squared Error (MSE)
    mse = np.mean((y - y_pred)**2)
    # R-squared value
    ss_total = np.sum((y - np.mean(y))**2)
    ss_res = np.sum((y - y_pred)**2)
    r_squared = 1 - (ss_res / ss_total)
    return mse, r_squared
```

```python
def modelFit(X, Y, n):
    degree = n # Degree of the polynomial
    coefficients = np.polyfit(X, Y, degree)
    # Step 3: Use the coefficients to predict y-values (polynomial function)
    # Generate polynomial function based on the coefficients
    polynomial = np.poly1d(coefficients)
    # Generate predicted y-values based on the fitted polynomial
    y_pred = polynomial(X)
    # Step 4: Plotting the results

    print('Least Squares Method:')
    mse_leastSquares, r_squared_leastSquares = evaluateFit(Y, m*X + c)
    print(f'Mean Squared Error: {mse_leastSquares:.4f}')
    print(f'R-squared: {r_squared_leastSquares:.4f}\n')

    print('Polynomial Fit:')
    mse_polyFit, r_squared_polyFit = evaluateFit(Y, y_pred)
    print(f'Mean Squared Error: {mse_polyFit:.4f}')
    print(f'R-squared: {r_squared_polyFit:.4f}\n')
    
    plt.scatter(X, Y, color='blue', label='Data points') # Original data points
    plt.plot(X, m*X + c, 'g-', label = 'Least Squares Method')
    plt.plot(X, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```

## Challenge A:

```python
X = np.random.rand(50)*10
X = np.sort(X)
Y = np.random.rand(50)*10

A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

degree = 3 # Degree of the polynomial
coefficients = np.polyfit(X, Y, degree)
# Step 3: Use the coefficients to predict y-values (polynomial function)
# Generate polynomial function based on the coefficients
polynomial = np.poly1d(coefficients)
# Generate predicted y-values based on the fitted polynomial
y_pred = polynomial(X)
# Step 4: Plotting the results

print('Least Squares Method:')
mse_leastSquares, r_squared_leastSquares = evaluateFit(Y, m*X + c)
print(f'Mean Squared Error: {mse_leastSquares:.4f}')
print(f'R-squared: {r_squared_leastSquares:.4f}\n')

print('Polynomial Fit:')
mse_polyFit, r_squared_polyFit = evaluateFit(Y, y_pred)
print(f'Mean Squared Error: {mse_polyFit:.4f}')
print(f'R-squared: {r_squared_polyFit:.4f}\n')
    
plt.scatter(X, Y, color='blue', label='Data points') # Original data points
plt.plot(X, m*X + c, 'g-', label = 'Least Squares Method')
plt.plot(X, y_pred, color='red', label=f'Polynomial fit of degree {degree}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

if mse_polyFit < mse_leastSquares and r_squared_polyFit > r_squared_leastSquares:
    print('Polynomial Method is a better fit.')
else:
    print('Least Squared Method is a better fit.')
```

```python
# Create Vandermonde matrix for a polynomial of degree 3
degree = 3
V = np.vander(X, N=degree+1, increasing=True)

# V.T * V
VT_V = np.dot(V.T, V)
# V.T * y
VT_y = np.dot(V.T, Y)
# Solve for a (coefficients)
coefficients = np.linalg.solve(VT_V, VT_y)
y_pred = np.dot(V, coefficients)

plt.scatter(X, Y, color='blue', label='Data points') # Original data points
plt.plot(X, y_pred, color='red', label=f'Vandermonde Matrix')
# Fitted curve
plt.title('Polynomial Fit - Vandermonde')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
